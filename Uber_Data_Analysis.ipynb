```python
import pandas as pd
```

```python
df=pd.read_csv('uber_data.csv')
df.head()
```

```python
df.info()
```

# Datetime_Dim Table

```python
# Convert DateTime Columns to Datatime Format
df['tpep_pickup_datetime']=pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime']=pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
datetime_dim=df[['tpep_pickup_datetime','tpep_dropoff_datetime']].drop_duplicates().reset_index(drop=True)
```

```python
# pickup Datetime
datetime_dim['pickup_hours']=datetime_dim['tpep_pickup_datetime'].dt.hour
datetime_dim['pickup_days']=datetime_dim['tpep_pickup_datetime'].dt.day
datetime_dim['pickup_month']=datetime_dim['tpep_pickup_datetime'].dt.month
datetime_dim['pickup_year']=datetime_dim['tpep_pickup_datetime'].dt.year
datetime_dim['pickup_week_day']=datetime_dim['tpep_pickup_datetime'].dt.weekday
datetime_dim
```

```python
# dropoff Datetime
datetime_dim['dropoff_hours']=datetime_dim['tpep_dropoff_datetime'].dt.hour
datetime_dim['dropoff_days']=datetime_dim['tpep_dropoff_datetime'].dt.day
datetime_dim['dropoff_month']=datetime_dim['tpep_dropoff_datetime'].dt.month
datetime_dim['dropoff_year']=datetime_dim['tpep_dropoff_datetime'].dt.year
datetime_dim['dropoff_week_day']=datetime_dim['tpep_dropoff_datetime'].dt.weekday
datetime_dim
```

```python
#create datetime Id
datetime_dim['datetime_id']=datetime_dim.index
datetime_dim
```

```python
# Arrange All columns 
datetime_dim=datetime_dim[['datetime_id','tpep_pickup_datetime', 'tpep_dropoff_datetime', 'pickup_hours',
                          'pickup_days', 'pickup_month', 'pickup_year', 'pickup_week_day',
                          'dropoff_hours', 'dropoff_days', 'dropoff_month', 'dropoff_year',
                          'dropoff_week_day']]
datetime_dim
```

# DropOff_Dim Table

```python
drop_off_dim=df[['dropoff_longitude','dropoff_latitude']].drop_duplicates().reset_index(drop=True)
drop_off_dim['dropoff_id']=drop_off_dim.index
drop_off_dim=drop_off_dim[['dropoff_id','dropoff_longitude','dropoff_latitude']]
drop_off_dim
```

# PickUp_Dim Table

```python
pick_up_dim=df[['pickup_longitude','pickup_latitude']].drop_duplicates().reset_index(drop=True)
pick_up_dim['pickup_id']=pick_up_dim.index
pick_up_dim=pick_up_dim[['pickup_id','pickup_longitude','pickup_latitude']]
pick_up_dim
```

# Payment_Dim Table

```python
payment_type_dim=df[['payment_type']].drop_duplicates().reset_index(drop=True)
payment_type_dim['payment_type_id']=payment_type_dim.index
payment_type_dim['payment_type_name']=payment_type_dim['payment_type'].map({1:'Credit Card',2:'Cash',3:'No Charge',4:'Dispute'})
payment_type_dim=payment_type_dim[['payment_type_id','payment_type','payment_type_name']]
payment_type_dim
```

# Rate_code_Dim Table

```python
Rate_code_dim=df[['RatecodeID']].drop_duplicates().reset_index(drop=True)
Rate_code_dim['Rate_code_id']=Rate_code_dim.index
Rate_code_dim['Rate_code_Name']=Rate_code_dim['RatecodeID'].map({1:'Standard',2:'JFK',3:'Newark',4:'Westchester',5:'Negotiated fare',6:'Group ride'})
Rate_code_dim=Rate_code_dim[['Rate_code_id','RatecodeID','Rate_code_Name']]
Rate_code_dim
```
